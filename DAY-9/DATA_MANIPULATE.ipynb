```python
import numpy as np
```

```python
salaries = np.array([
    30000,
    45000,
    50000,
    60000,
    80000
])
```

```python
print(
    "Total:",
    np.sum(salaries)
)

print(
    "Average:",
    np.mean(salaries)
)

print(
    "Highest:",
    np.max(salaries)
)

print(
    "Lowest:",
    np.min(salaries)
)
```

```python
new_salaries = salaries * 1.10

print(new_salaries)
```

```python
high_earners = salaries[
    salaries > 50000
]

print(high_earners)
```
